```python
from collections import defaultdict
makes=defaultdict(int)

f=open('./cars_2014.csv','r')
for i in f :
    l,_=i.strip('\n').split(',')
    makes[l]+=1
for key,values in makes.items() :

    print(f'{key}={values}')
```

```python
from itertools import *

f=open('./cars_2014.csv','r')
next(f)
l=groupby(f,key= lambda x : (x.strip('\n').split(','))[0])
for i in l :
    print(i[0],len(list(i[1])))
```
